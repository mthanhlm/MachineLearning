```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Nhập dữ liệu
data = pd.read_csv('/Users/daokhanh/Desktop/Machine Learning/Ex_1/california_housing_train.csv')
m = len(data)
X = data['x'].values.reshape(m, 1)
y = data['y'].values.reshape(m, 1)

# Visualize dữ liệu
plt.scatter(X, y, color='red')
plt.title('Visualize dữ liệu')
plt.show()
```

```python
# Thêm cột 1 vào giá trị X
X = np.hstack((np.ones((m, 1)), X))
print('Giá trị của X sau khi thêm cột :')
print(X)
```

```python

# Hàm Gradient descent
def gradient_descent(X, y, theta, alpha, iteration):
    m = len(y)
    J_history = np.zeros((iteration, 1))

    for i in range(iteration):
        theta = theta - (alpha/m) * (X.T @ (X @ theta - y))
        J_history[i] = compute_cost(X, y, theta)

    return theta, J_history

```

```python
# Hàm tính Cost J 
def compute_cost(X, y, theta):
    m = len(y)
    J = (1/(2*m)) * np.sum((X @ theta - y)**2)
    return J
```

```python
# Khởi tạo giá trị theta ,learning rate(alpha) và số vòng lặp
theta = np.zeros((2, 1))
alpha = 0.0001
iteration = 10
```

```python
# Chạy Gradient Descent
theta, J_history = gradient_descent(X, y, theta, alpha, iteration)

# Tác động của số lượng vòng lặp và các giá trị của learning rate vào hàm cost
plt.plot(range(len(J_history)), J_history,color = 'black')
plt.title('Tác động của số lượng vòng lặp và các giá trị của learning rate vào hàm cost')
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.show()

```

```python
# Đường thẳng cuối cùng
plt.scatter(X[:, 1], y, color = 'red')
plt.plot(X[:, 1], X @ theta, color='black')
plt.title('Đường thẳng cuối cùng')
plt.show()
```
